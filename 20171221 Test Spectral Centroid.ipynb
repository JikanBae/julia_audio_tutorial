```julia
using LibSndFile
include("audioutil.jl");
```

```julia
snd = load("recorder.flac");
x = snd.data;
fs = snd.samplerate;
window_size = 1024;
hop_size = 256;
sc, t_sc = spectral_centroid_moving(x, fs, window_size, hop_size);
rmss, t_rmss = rms_moving(x, fs, window_size, hop_size);
```

```julia
plot(t_sc, sc);
ylim(400, 700);
grid()
```

```julia
plot(t_rmss, rmss);
grid()
```

```julia
plot(t_sc, sc .* (rmss / maximum(rmss)));
grid();
```
